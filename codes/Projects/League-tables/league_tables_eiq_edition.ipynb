```python
#broker = pd.read_excel(r'C:\Users\Pawandeep Kaur\RepuTex\Company - Documents\10 Python\Codes\League Tables\broker-market-data.xlsx',sheet_name='broker share',header=1#[0, 1])
broker_fp = (
    Path.home()
    / "RepuTex"
    / "Company - Documents"
    / "10 Python"
    / "Codes"
    / "League Tables"
    / "broker-market-data.xlsx"
)

broker = pd.read_excel(
    broker_fp,
    sheet_name="broker share",
    header=1
)
# 1. Identify transaction value columns
tx_cols = [col for col in broker.columns if "Transaction value" in str(col)]

# 2. Identify percentage columns manually or by name
pct_cols = [col for col in broker.columns if "%" in str(col) or "share" in str(col).lower() or "percent" in str(col).lower()]

# 3. Clean transaction value columns and convert to numeric
for col in tx_cols:
    broker[col] = (
        broker[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )
    broker[col] = pd.to_numeric(broker[col], errors='coerce').round(1)

# 4. Find numeric columns excluding tx and pct
num_cols = broker.select_dtypes(include='number').columns
other_num_cols = [col for col in num_cols if col not in tx_cols and col not in pct_cols]

# 5. Round other numeric columns to 1 decimal
broker[other_num_cols] = broker[other_num_cols].round(1)

# 6. Format transaction value columns with $ and 1 decimal
for col in tx_cols:
    broker[col] = broker[col].map(lambda x: f'${x:,.1f}' if pd.notna(x) else '')

# 7. Format percentage columns back to % text
for col in pct_cols:
    broker[col] = broker[col].map(lambda x: f'{x*100:.1f}%' if pd.notna(x) else '')
broker.head(10)
```

```python
#market = pd.read_excel(r'C:\Users\Pawandeep Kaur\RepuTex\Company - Documents\10 Python\Codes\League Tables\broker-market-data.xlsx',sheet_name='market value new',header=1)
market_fp = (
    Path.home()
    / "RepuTex"
    / "Company - Documents"
    / "10 Python"
    / "Codes"
    / "League Tables"
    / "broker-market-data.xlsx"
)

market = pd.read_excel(
    market_fp,
    sheet_name="market value new",
    header=1
)
# 1. Currency transaction columns only
tx_cols = [
    col for col in market.columns
    if "Transaction value" in str(col) and "%" not in str(col)
]

# 2. Percentage columns
pct_cols = [
    col for col in market.columns
    if "%" in str(col) or "share" in str(col).lower() or "percent" in str(col).lower()
]

# 3. Clean and convert currency transaction columns
for col in tx_cols:
    market[col] = (
        market[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )
    market[col] = pd.to_numeric(market[col], errors='coerce').round(1)

# 4. Convert percentage columns to numeric if needed, but do not round them
for col in pct_cols:
    market[col] = (
        market[col]
        .astype(str)
        .str.replace('%', '', regex=False)
        .str.strip()
    )
    market[col] = pd.to_numeric(market[col], errors='coerce')

    # if values are like 37 instead of 0.37, convert to proportion
    if market[col].dropna().gt(1).any():
        market[col] = market[col] / 100

# 5. Round other numeric columns to 1 decimal
num_cols = market.select_dtypes(include='number').columns
other_num_cols = [col for col in num_cols if col not in tx_cols and col not in pct_cols]
market[other_num_cols] = market[other_num_cols].round(1)

# 6. Format currency columns
for col in tx_cols:
    market[col] = market[col].map(lambda x: f'${x:,.1f}' if pd.notna(x) else '')

# 7. Format percentage columns back
for col in pct_cols:
    market[col] = market[col].map(lambda x: f'{x*100:.1f}%' if pd.notna(x) else '')

market.head(15)

```
